```python
from Bio import SeqIO
import pandas as pd

def fasta_to_dataframe(fasta_path):
    records = []
    for record in SeqIO.parse(fasta_path, "fasta"):
        records.append({"id": record.id, "sequence": str(record.seq)})
    return pd.DataFrame(records)

df = fasta_to_dataframe("data/zz/train.fasta")
df.to_csv("data/zz/train.csv", index=False)
```

```python
def aa_to_int(sequence, padding_length=None):
    """
    Converts an amino acid sequence to a list of integers (1-based index).
    Optionally pads the sequence to a fixed length with 0s.
    """
    aa_map = {
        'A': 1, 'C': 2, 'D': 3, 'E': 4, 'F': 5,
        'G': 6, 'H': 7, 'I': 8, 'K': 9, 'L': 10,
        'M': 11, 'N': 12, 'P': 13, 'Q': 14, 'R': 15,
        'S': 16, 'T': 17, 'V': 18, 'W': 19, 'Y': 20
    }

    encoded = [aa_map.get(aa, 0) for aa in sequence.upper()]  # unknowns → 0

    if padding_length:
        if len(encoded) > padding_length:
            encoded = encoded[:padding_length]
        else:
            encoded += [0] * (padding_length - len(encoded))

    return encoded
```

```python
import os
from collections import defaultdict

def parse_clstr_file(clstr_path):
    """
    Parses a .clstr file and returns a list of clusters,
    each cluster is a list of sequence names.
    """
    clusters = []
    with open(clstr_path, "r") as f:
        current = []
        for line in f:
            if line.startswith(">Cluster"):
                if current:
                    clusters.append(current)
                current = []
            else:
                seq_id = line.split(">")[1].split("...")[0]
                current.append(seq_id)
        if current:
            clusters.append(current)
    return clusters

def compute_upsampling(clusters, target_size=32, max_factor=50):
    """
    Assign upsampling factors to each sequence based on target cluster size.
    
    Args:
        clusters: list of clusters (each cluster is a list of sequence IDs)
        target_size: the target cluster size to normalize upsampling
        max_factor: optional cap to avoid runaway inflation of upsampling

    Returns:
        upsampling_map: dict mapping sequence ID → upsampling factor
    """
    upsampling_map = {}
    upsampling_factors = []

    for cluster in clusters:
        cluster_size = len(cluster)
        factor = max(1, target_size // cluster_size)
        factor = min(factor, max_factor)

        for seq_id in cluster:
            upsampling_map[seq_id] = factor
            upsampling_factors.append(factor)

    # Logging
    import numpy as np
    print("Upsampling factor stats:")
    print(f"  Min: {min(upsampling_factors)}")
    print(f"  Max: {max(upsampling_factors)}")
    print(f"  Mean: {np.mean(upsampling_factors):.2f}")
    print(f"  Total synthetic samples after upsampling: {sum(upsampling_factors)} (original: {sum(len(c) for c in clusters)})")

    return upsampling_map
```

```python
import os
import pandas as pd
import tensorflow as tf
from collections import defaultdict

def assign_bin(factor):
    if factor <= 1:
        return 'bin_1'
    elif factor <= 2:
        return 'bin_2'
    elif factor <= 4:
        return 'bin_4'
    elif factor <= 8:
        return 'bin_8'
    elif factor <= 16:
        return 'bin_16'
    else:
        return 'bin_32'


def write_clustered_tfrecords(csv_file, upsampling_map, output_dir):
    df = pd.read_csv(csv_file)
    os.makedirs(output_dir, exist_ok=True)

    # Group sequence IDs into bins
    bin_groups = defaultdict(list)
    for seq_id, factor in upsampling_map.items():
        bin_name = assign_bin(factor)
        bin_groups[bin_name].append((seq_id, factor))

    # Write a TFRecord file for each bin
    for bin_name, entries in bin_groups.items():
        filename = os.path.join(output_dir, f"{bin_name}.tfrecords")
        with tf.io.TFRecordWriter(filename) as writer:
            for seq_id, factor in entries:
                seq_row = df[df['id'] == seq_id]
                if seq_row.empty:
                    continue
                seq = seq_row['sequence'].values[0]
                encoded = aa_to_int(seq)

                for _ in range(factor):
                    feature = {
                        "label": tf.train.Feature(int64_list=tf.train.Int64List(value=[0])),  # Dummy label
                        "sequence": tf.train.Feature(int64_list=tf.train.Int64List(value=encoded))
                    }
                    example = tf.train.Example(features=tf.train.Features(feature=feature))
                    writer.write(example.SerializeToString())
        print(f"{bin_name}: {len(entries)} sequences")
```

```python
import random

def sample_real_sequences_proportional(clusters, total_target=500, max_per_cluster=20, seed=42):
    """
    Samples real sequence IDs proportionally by cluster size, capped by max_per_cluster.
    
    Args:
        clusters (list of list of str): List of clusters, each a list of sequence IDs.
        total_target (int): Approximate total number of sequences to sample.
        max_per_cluster (int): Cap on how many sequences to sample per cluster.
        seed (int): Random seed for reproducibility.
    
    Returns:
        sampled_ids (list of str): List of sampled sequence IDs.
    """
    random.seed(seed)
    total_seqs = sum(len(c) for c in clusters)
    sampled_ids = []

    for cluster in clusters:
        proportion = len(cluster) / total_seqs
        n = min(max_per_cluster, max(1, int(proportion * total_target)))
        sampled_ids.extend(random.sample(cluster, min(n, len(cluster))))

    # Adjust if we overshot or undershot the target
    if len(sampled_ids) > total_target:
        sampled_ids = random.sample(sampled_ids, total_target)
    elif len(sampled_ids) < total_target:
        # Top up randomly (optional)
        remaining = total_target - len(sampled_ids)
        flat = [sid for cluster in clusters for sid in cluster if sid not in sampled_ids]
        sampled_ids.extend(random.sample(flat, min(remaining, len(flat))))

    return sampled_ids
```

```python
clusters = parse_clstr_file("data/zz/clustered_train_80.fasta.clstr")
upsample_map = compute_upsampling(clusters)
```

```python
sampled_seq_ids = sample_real_sequences_proportional(clusters, total_target=500)
```

```python
df = pd.read_csv("data/zz/train.csv")
sample_seqs = df[df['id'].isin(sampled_seq_ids)].reset_index(drop=True) # For comparing embeddings vs generated 
```
